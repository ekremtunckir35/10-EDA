The Census Income dataset has 48,842 entries. Each entry contains the following information about an individual:

- **salary (target feature/label):** whether or not an individual makes more than $50,000 annually. (<= 50K, >50K)
- **age:** the age of an individual. (Integer greater than 0)
- **workclass:** a general term to represent the employment status of an individual. (Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked)
- **fnlwgt: ** this is the number of people the census believes the entry represents. People with similar demographic characteristics should have similar weights.  There is one important caveat to remember about this statement. That is that since the CPS sample is actually a collection of 51 state samples, each with its own probability of selection, the statement only applies within state.(Integer greater than 0)
- **education:** the highest level of education achieved by an individual. (Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool.)
- **education-num:** the highest level of education achieved in numerical form. (Integer greater than 0)
- **marital-status:** marital status of an individual. Married-civ-spouse corresponds to a civilian spouse while Married-AF-spouse is a spouse in the Armed Forces. Married-spouse-absent includes married people living apart because either the husband or wife was employed and living at a considerable distance from home (Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse)
- **occupation:** the general type of occupation of an individual. (Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces)
- **relationship:** represents what this individual is relative to others. For example an individual could be a Husband. Each entry only has one relationship attribute. (Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried)
- **race:** Descriptions of an individual’s race. (White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black)
- **sex:** the biological sex of the individual. (Male, female)
- **capital-gain:** capital gains for an individual. (Integer greater than or equal to 0)
- **capital-loss:** capital loss for an individual. (Integer greater than or equal to 0)
- **hours-per-week:** the hours an individual has reported to work per week. (continuous)
- **native-country:** country of origin for an individual (United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands)



Nüfus Sayımı Gelir veri setinde 48.842 girdi bulunmaktadır. Her girdi bir birey hakkında aşağıdaki bilgileri içerir:

- **maaş (hedef özellik/etiket):** bir bireyin yıllık 50.000 $'dan fazla kazanıp kazanmadığı. (<= 50K, >50K)
- yaş:** bir bireyin yaşı. (0'dan büyük tamsayı)
- İş sınıfı:** bir bireyin istihdam durumunu temsil eden genel bir terimdir. (Özel, Kendi hesabına çalışmayan, Kendi hesabına çalışan, Federal hükümet, Yerel hükümet, Eyalet hükümeti, Ücretsiz, Hiç çalışmamış)
- fnlwgt:** bu, nüfus sayımının girdinin temsil ettiğine inandığı kişi sayısıdır. Benzer demografik özelliklere sahip kişiler benzer ağırlıklara sahip olmalıdır.  Bu ifade hakkında hatırlanması gereken önemli bir uyarı vardır. O da, CPS örnekleminin aslında her biri kendi seçilme olasılığına sahip 51 eyalet örnekleminden oluşması nedeniyle, ifadenin yalnızca eyalet içinde geçerli olduğudur.(Integer greater than 0)
- **eğitim:** bir birey tarafından ulaşılan en yüksek eğitim seviyesi. (Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool)
- eğitim-num:** sayısal olarak elde edilen en yüksek eğitim seviyesi. (0'dan büyük tamsayı)
- medeni durum:** bir bireyin medeni durumu. Evli-sivil-eş sivil bir eşe karşılık gelirken, Evli-AF-eş Silahlı Kuvvetlerdeki bir eştir. Evli-eş-yok, karı veya kocadan birinin çalışması nedeniyle ayrı yaşayan ve evden önemli ölçüde uzakta yaşayan evli kişileri kapsamaktadır (Evli-sivil-eş, Boşanmış, Hiç evlenmemiş, Ayrılmış, Dul, Evli-eş-yok, Evli-AF-eş)
- Meslek:** bir bireyin genel meslek türü. (Teknik-destek, Zanaat-tamir, Diğer-hizmet, Satış, Yönetici-yönetim, Profesyonel-uzmanlık, İşleyiciler-temizlikçiler, Makine-op-inspct, Yönetici-büro, Çiftçilik-balıkçılık, Nakliye-hareket, Özel-ev-hizmet, Koruyucu-hizmet, Silahlı Kuvvetler)
- ilişki:** bu bireyin diğerlerine göre ne olduğunu temsil eder. Örneğin bir birey Koca olabilir. Her girişin yalnızca bir ilişki niteliği vardır. (Eş, Kendi-çocuğu, Koca, Aileden değil, Diğer-akraba, Evli değil)
- Irk:** Bir bireyin ırkının tanımı. (Beyaz, Asya-Pasifik-İskandinav, Amerikan-Hint-Eskimo, Diğer, Siyah)
- **seks:** bireyin biyolojik cinsiyeti. (Erkek, kadın)
- sermaye kazancı:** bir birey için sermaye kazancı. (0'a eşit veya 0'dan büyük tamsayı)
- sermaye kaybı:** bir birey için sermaye kaybı. (0'a eşit veya 0'dan büyük tamsayı)
- Haftalık çalışma saati:** bir bireyin haftada çalıştığını bildirdiği saat. (sürekli)
- anavatan-ülke: ** Bir bireyin menşe ülkesi (Birleşik Devletler, Kamboçya, İngiltere, Porto-Riko, Kanada, Almanya, Dış-ABD (Guam-USVI- vb.), Hindistan, Japonya, Yunanistan, Güney, Çin, Küba, İran, Honduras, Filipinler, İtalya, Polonya, Jamaika, Vietnam, Meksika, Portekiz, İrlanda, Fransa, Dominik Cumhuriyeti, Laos, Ekvador, Tayvan, Haiti, Kolombiya, Macaristan, Guatemala, Nikaragua, İskoçya, Tayland, Yugoslavya, El-Salvador, Trinad&Tobago, Peru, Hong, Holanda-Hollanda)

### Boxplot

- Kutu grafiği yaş verilerini özetlemek için kullanılır.<br>
- Kutunun alt ve üst sınırları, verinin ilk (Q1) ve üçüncü çeyreklik (Q3) değerlerini gösterir. Bu durumda Q1 yaklaşık 30, Q3 ise yaklaşık 45 civarında.<br>
- Kutunun içindeki çizgi medyanı (Q2) gösterir ve bu değer yaklaşık 35 civarında.<br>
- Kutunun dışındaki çizgiler ("bıyıklar" olarak adlandırılır) veri setindeki minimum ve maksimum değerleri gösterir, ancak uç değerleri (outliers) hariç tutar. Bu grafikte bıyıkların sınırları yaklaşık 20 ile 80 arasında.<br>
- Kutunun dışında kalan noktalar ise uç değerlerdir (outliers).<br>
- Bu, veri setinde genel trendin dışında kalan, istatistiksel olarak anormal olan değerlerdir. Bu grafikte, 80 yaşın üzerindeki değerler uç değer olarak görülüyor.


### Histogram 

- Histogram yaş dağılımının frekansını gösterir.<br>
- X ekseninde yaş grupları ve Y ekseninde bu yaş grubundaki kişi sayısı gösterilir.<br>
- Bu grafikte, en fazla kişi sayısının 30-40 yaş arasında olduğu görülüyor. Bu yaş aralığında yaklaşık 6000 kişi bulunuyor.<br>
- Yaş dağılımı sağa doğru azalan bir yapıya sahip, yani daha genç yaş gruplarında daha fazla kişi var ve yaş ilerledikçe kişi sayısı azalıyor.<br>
- 20-30 yaş aralığında da yüksek bir frekans var, ardından 40-50 yaş aralığında bir düşüş başlıyor.
60 yaş ve üzeri gruplarda kişi sayısı oldukça azalmış durumda.
